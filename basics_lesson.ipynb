## Listen und Dictionaries

```python
# Liste ausgeben lassen
list_names
```

```python
# Erste Position der Liste ausgeben lassen
list_names[0]
```

```python
# Es können keine Positionen ausgegeben werden, die nicht existieren
list_names[20]
```

```python
# Länge der Liste ausgeben lassen
len(list_names)
```

```python
# letzte Position der Liste ausgeben lassen
list_names[-1]
```

```python
# mehrere Positionen aus der Liste ausgeben lassen
list_names[0:3]
```

```python
# Slicing, alles vor einer bestimmten Position ausgeben lassen
list_names[:2]
```

```python
# Slicing, alles nach einer bestimmten Position ausgeben lassen
list_names[2:]
```

## for-Schleifen

```python
# alle Positionen der Liste ausgeben lassen als for-Schleife
for person in list_names:
    print(person)
```

```python
# alle Positionen der Liste ausgeben lassen ohne for-Schleife
list_names
```

```python
# den Positionen der Liste noch eine Zeichenkette hinzufügen
for person in list_names:
    if person == "Lovelace":
        print(person + " did awesome stuff")
    else:
        print(person)
```
