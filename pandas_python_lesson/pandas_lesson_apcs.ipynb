# APCs plotting - pandas lesson

```python
import pandas as pd
import urllib.request
```

```python
# Variablen definieren
url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
apc_file = "openapc.csv"
```

```python
# Download der Daten als CSV-Datei
urllib.request.urlretrieve(url, apc_file)
```

```python
# Einlesen der CSV-Datei in einen pandas DataFrame
df_apcs = pd.read_csv(apc_file)
```
