# Metadaten per DOI abfragen - DataCite lesson

```python
import urllib.request
import json
```

```python
full_url = "https://api.datacite.org/application/vnd.datacite.datacite+json/10.5281/zenodo.5563162"
```

```python
# Ausgabe der URL einlesen
response_doi_json_dataset = urllib.request.urlopen(full_url).read()
```

```python
# Response wird zu JSON-Daten umgewandelt
doi_dataset = json.loads(response_doi_json_dataset)
```
